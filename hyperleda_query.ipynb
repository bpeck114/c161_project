```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Hyperleda Query

```python
plt.rcParams.update({'font.size': 25})
plt.style.use("dark_background")
```

```python
df = pd.read_csv('hyperleda_calc.csv').dropna()
df
```

```python
distance_mpc = df['distance_mpc']
vel_gsr =  df['v_gsr']
```

```python
plt.figure(figsize = (25,13))
plt.scatter(distance_mpc,vel_gsr)
plt.title('Raw Data with outliers')
plt.xlabel('Distance(Mpc)')
plt.ylabel('Vel(km/s)')
```

```python
df_filtered = df.where(df.where(df['distance_mpc']<200).dropna()['v_gsr']<20000)
distance_mpc_filtered = df_filtered['distance_mpc']
vel_gsr_filtered =  df_filtered['v_gsr']
df_filtered
```

```python
plt.figure(figsize = (25,13))
plt.scatter(distance_mpc_filtered,vel_gsr_filtered)
#plt.ylim(-1000, 20000)
plt.title('Raw Data without outliers')
plt.xlabel('Distance(Mpc)')
plt.ylabel('Vel(km/s)')
plt.show()
```
